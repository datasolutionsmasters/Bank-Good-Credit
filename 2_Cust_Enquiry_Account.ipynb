```python
import pandas as pd
import numpy as np
```

# Enquiry File Transformation

```python
df = pd.read_csv('1_Cust_Enquiry.csv')
df.head()
```

```python
df.columns
```

```python
df1 = df.loc[:,['customer_no','enq_amt']]
#df1[df1.customer_no==1].count()
df1_back = df1.copy()
```

```python
Cust_group = df1.groupby(['customer_no'])
#Cust_group.count()
```

```python
df_amt = pd.DataFrame(Cust_group.sum()).reset_index()          
df_amt.columns = ['customer_no','total_enq_amt']
df_amt.head()
df_amt.shape
```

```python
df_count = pd.DataFrame(Cust_group.count()).reset_index()          
df_count.columns = ['customer_no','total_enq_count']
df_count.head()
df_count.shape
```

```python
# Inner Join
df_final = pd.merge(df_amt, df_count, on='customer_no', how='inner')
df_final.head()
df_final.shape
```

```python
df_final.to_csv('2_Cust_enquiry_Final.csv',index=False)
```

# Account File transformation

```python
df_a = pd.read_csv('1_Cust_Account.csv')
df_a.head()
```

```python
df_a.columns
```

```python
ColList = ['customer_no','high_credit_amt', 'cur_balance_amt', 'amt_past_due', 'creditlimit','cashlimit']
df1_a = df_a.loc[:,ColList]
df1_a.head()
```

```python
df1_a.fillna(0,inplace=True)
df1_a.head()
```

```python
Cust_group_a = df1_a.groupby(['customer_no'])
#Cust_group_a.head()
```

```python
df_a_1 = pd.DataFrame(Cust_group_a.sum()).reset_index()          
df_a_1.columns = ['customer_no','tot_high_credit_amt','tot_cur_balance_amt','tot_amt_past_due','tot_creditlimit','tot_cashlimit']
df_a_1.head()
df_a_1.shape
```

```python
df_a_1.to_csv('2_Cust_Account_Final.csv',index=False)
```
